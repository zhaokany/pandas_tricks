```python
idx = pd.IndexSlice
```

```python
df_ex.loc[idx[("C"):,("D","f"):]]
```

```python
df_ex.loc[idx[:"A",lambda x:x.sum()>0]]
```

```python
# from_tuple
np.random.seed(0)
df = pd.read_csv("data/learn_pandas.csv", usecols=["School", "Grade", "Name", "Gender", "Weight", "Transfer"])
df_ = df.copy()

my_tuple1 = [(x,y) for x in list("ABCD") for y in df_.Gender.unique()]
multi_index = pd.MultiIndex.from_tuples(
    my_tuple1, names=("School", "Gender")
)
df_.Grade = df_.Grade.replace(
    ["Freshman", "Senior", "Sophomore", "Junior"], [1,4,2,3]
)
my_tuple2 = [(x,y) for x in ["Height", "Weight"] for y in df_.Grade.unique()]
multi_column = pd.MultiIndex.from_tuples(
    my_tuple2, names=("Indicator", "Grade")    
)
df_multi = pd.DataFrame(
    np.c_[(np.random.randn(8,4)*5 + 163).tolist(),
         (np.random.rand(8,4)*5 + 65).tolist()],
    index=multi_index,
    columns=multi_column).round(1).sort_values(["Indicator", "Grade"], axis=1)

df_multi
```

```python
# from_arrays
np.random.seed(0)
df = pd.read_csv("data/learn_pandas.csv", usecols=["School", "Grade", "Name", "Gender", "Weight", "Transfer"])
df_ = df.copy()

my_array1 = [list("AABBCCDD"), list(df_.Gender.unique())*4]
multi_index = pd.MultiIndex.from_arrays(
    my_array1, names=("School", "Gender")
)
df_.Grade = df_.Grade.replace(
    ["Freshman", "Senior", "Sophomore", "Junior"], [1,4,2,3]
)
my_array2 = [["Height"]*4 + ["Weight"]*4, list(df_.Grade.unique())*2]
multi_column = pd.MultiIndex.from_arrays(
    my_array2, names=("Indicator", "Grade")    
)
df_multi = pd.DataFrame(
    np.c_[(np.random.randn(8,4)*5 + 163).tolist(),
         (np.random.rand(8,4)*5 + 65).tolist()],
    index=multi_index,
    columns=multi_column).round(1).sort_values(["Indicator", "Grade"], axis=1)

df_multi
```
