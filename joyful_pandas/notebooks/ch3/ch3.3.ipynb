```python
np.random.seed(0)
L1,L2,L3=["A","B"],["a","b"],["alpha","beta"]
mul_index1 = pd.MultiIndex.from_product([L1,L2,L3], names=("Upper", "Lower", "Extra"))
L4,L5,L6=["C","D"],["c","d"],["cat","dog"]
mul_index2 = pd.MultiIndex.from_product([L4,L5,L6], names=("Big", "Small", "Other"))

df_ex = pd.DataFrame(np.random.randint(-9,10,(8,8)), index=mul_index1, columns=mul_index2)
df_ex.head()
```

```python
df_ex.swaplevel(0,2,axis=1).head()
```

```python
df_ex.reorder_levels([2,0,1],axis=0).head()
```

```python
df_ex.droplevel(1, axis=1).head()
```

```python
df_ex.droplevel([0,1],axis=0).head()
```

```python
df_ex.rename_axis(index={"Upper": "Changed_row"},
                  columns={"Other": "Changed_col"}).head()
```

```python
df_ex.rename(columns={"cat": "not_cat"}, level=2)
```

```python
df_ex.rename(index=lambda x:str.upper(x), level=2).head()
```

```python
# Ex3-5
```

```python
df_ex.rename_axis(["Changed_row", "Lower", "Extra"], axis=0).rename_axis(["Big", "Small", "Changed_col"], axis=1)
```

```python
new_values = iter(list("abcdefgh"))
df_ex.rename(index=lambda x:next(new_values), level=2).head()
```

```python
df_temp = df_ex.copy()
new_idx = df_temp.index.map(lambda x: (x[0], x[1], str.upper(x[2])))
df_temp.index = new_idx
df_temp.head()
```

```python
df_temp = df_ex.copy()
new_idx = df_temp.index.map(lambda x: (x[0]+"-"+x[1]+"-"+x[2]))
df_temp.index = new_idx
df_temp.head()
```

```python
new_idx = df_temp.index.map(lambda x: tuple(x.split("-")))
df_temp.index = new_idx
df_temp.head()
```

```python
df_temp = df_ex.copy()
new_col = df_temp.columns.map(lambda x: (x[0]+"-"+x[1]+"-"+x[2]))
df_temp.columns = new_col
df_temp.head()
```

```python
df_temp = df_ex.copy()
new_col = df_temp.columns.map(lambda x: (x[0]+"_"+x[2], x[1], x[2]))
df_temp.columns = new_col
df_temp.rename_axis(columns={"Big": "Big_Other"}).droplevel(2, axis=1)
```

```python
df_temp.rename(columns={"Big": "Big_Other"}, level=0).droplevel(2, axis=1).head()
```
